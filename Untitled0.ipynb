```python
import pandas as pd
import numpy as np

df = pd.read_csv('global_air_quality_dataset.csv')

print("First 5 rows:")
display(df.head())

print("\nShape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Fill missing numerical values with median
num_cols = df.select_dtypes(include=np.number).columns.tolist()
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill missing categorical values with mode
cat_cols = df.select_dtypes(include='object').columns.tolist()
for col in cat_cols:
    mode_val = df[col].mode()
    if not mode_val.empty:
        df[col] = df[col].fillna(mode_val[0])

# Convert date column if exists
date_cols = [col for col in df.columns if 'date' in col.lower() or 'time' in col.lower()]
if date_cols:
    df[date_cols[0]] = pd.to_datetime(df[date_cols[0]], errors='coerce')
    df['Year'] = df[date_cols[0]].dt.year
    df['Month'] = df[date_cols[0]].dt.month
elif 'Year' not in df.columns:
    df['Year'] = 2020
    df['Month'] = 1

print("Cleaned data shape:", df.shape)
display(df.head())
```

```python
# Check if AQI Category column already exists
print("Columns available:", df.columns.tolist())

# Auto-detect AQI value column
aqi_col = None
for col in df.columns:
    if 'aqi' in col.lower() and df[col].dtype in [np.float64, np.int64]:
        aqi_col = col
        break

print("AQI column detected:", aqi_col)

# Create AQI Category if not present
if 'AQI Category' not in df.columns:
    def categorize_aqi(aqi):
        if aqi <= 50: return 'Good'
        elif aqi <= 100: return 'Moderate'
        elif aqi <= 150: return 'Unhealthy for Sensitive Groups'
        elif aqi <= 200: return 'Unhealthy'
        elif aqi <= 300: return 'Very Unhealthy'
        else: return 'Hazardous'

    if aqi_col:
        df['AQI Category'] = df[aqi_col].apply(categorize_aqi)
        print("AQI Category created successfully!")
    else:
        print("No AQI column found!")
else:
    print("AQI Category column already exists!")

print("\nAQI Category counts:")
print(df['AQI Category'].value_counts())
```

```python
# Auto-detect AQI column
aqi_col = None
for col in df.columns:
    if 'aqi' in col.lower() and df[col].dtype in [np.float64, np.int64]:
        aqi_col = col
        break

print("Using AQI column:", aqi_col)

# Basic Statistics
print("\n===== BASIC STATISTICS =====")
print(f"Mean AQI:               {df[aqi_col].mean():.2f}")
print(f"Minimum AQI:            {df[aqi_col].min():.2f}")
print(f"Maximum AQI:            {df[aqi_col].max():.2f}")
print(f"Standard Deviation:     {df[aqi_col].std():.2f}")

# Auto-detect location column
loc_col = None
for col in df.columns:
    if any(x in col.lower() for x in ['country', 'city', 'location']):
        loc_col = col
        break

print(f"\nUsing location column: {loc_col}")

if loc_col:
    avg_aqi = df.groupby(loc_col)[aqi_col].mean().sort_values(ascending=False)
    print(f"\nHighest AQI {loc_col}: {avg_aqi.index[0]} ({avg_aqi.iloc[0]:.2f})")
    print(f"Lowest AQI  {loc_col}: {avg_aqi.index[-1]} ({avg_aqi.iloc[-1]:.2f})")

    print(f"\nTop 10 {loc_col}s by Average AQI:")
    display(avg_aqi.head(10).reset_index().rename(columns={aqi_col: 'Average AQI'}))
```

```python
# Auto-detect AQI column
aqi_col = None
for col in df.columns:
    if 'aqi' in col.lower() and df[col].dtype in [np.float64, np.int64]:
        aqi_col = col
        break

print("Using AQI column:", aqi_col)

# Basic Statistics
print("\n===== BASIC STATISTICS =====")
print(f"Mean AQI:               {df[aqi_col].mean():.2f}")
print(f"Minimum AQI:            {df[aqi_col].min():.2f}")
print(f"Maximum AQI:            {df[aqi_col].max():.2f}")
print(f"Standard Deviation:     {df[aqi_col].std():.2f}")

# Auto-detect location column
loc_col = None
for col in df.columns:
    if any(x in col.lower() for x in ['country', 'city', 'location']):
        loc_col = col
        break

print(f"\nUsing location column: {loc_col}")

if loc_col:
    avg_aqi = df.groupby(loc_col)[aqi_col].mean().sort_values(ascending=False)
    print(f"\nHighest AQI {loc_col}: {avg_aqi.index[0]} ({avg_aqi.iloc[0]:.2f})")
    print(f"Lowest AQI  {loc_col}: {avg_aqi.index[-1]} ({avg_aqi.iloc[-1]:.2f})")

    print(f"\nTop 10 {loc_col}s by Average AQI:")
    display(avg_aqi.head(10).reset_index().rename(columns={aqi_col: 'Average AQI'}))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Select numerical features only
num_cols_list = df.select_dtypes(include=np.number).columns.tolist()

# Remove AQI column from features
features_cols = [col for col in num_cols_list if 'aqi' not in col.lower()]
print("Features used:", features_cols)

X = df[features_cols].copy()
y = df['AQI Category'].copy()

# Encode target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n===== KNN CLASSIFICATION =====")
best_k = 3
best_acc = 0

for k in [3, 5, 7]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    acc = accuracy_score(y_test, knn.predict(X_test_scaled))
    print(f"KNN k={k} Accuracy: {acc:.4f}")
    if acc > best_acc:
        best_acc = acc
        best_k = k

print(f"\nBest k = {best_k} with accuracy = {best_acc:.4f}")

# Best KNN full report
knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(X_train_scaled, y_train)
y_pred_knn = knn_best.predict(X_test_scaled)

print("\nKNN Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))
print("\nKNN Classification Report:")
print(classification_report(y_test, y_pred_knn, target_names=le.classes_))

print("\n===== NAIVE BAYES CLASSIFICATION =====")
nb = GaussianNB()
nb.fit(X_train_scaled, y_train)
y_pred_nb = nb.predict(X_test_scaled)

nb_acc = accuracy_score(y_test, y_pred_nb)
print(f"Naive Bayes Accuracy: {nb_acc:.4f}")
print("\nNaive Bayes Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_nb))
print("\nNaive Bayes Classification Report:")
print(classification_report(y_test, y_pred_nb, target_names=le.classes_))

print("\n===== MODEL COMPARISON =====")
print(f"KNN Best Accuracy:      {best_acc:.4f}")
print(f"Naive Bayes Accuracy:   {nb_acc:.4f}")
if best_acc > nb_acc:
    print("Winner: KNN performed better!")
else:
    print("Winner: Naive Bayes performed better!")
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

print("===== K-MEANS CLUSTERING =====")

# Use same features as before (no AQI category)
X_cluster = df[features_cols].copy()

# Standardize
scaler2 = StandardScaler()
X_cluster_scaled = scaler2.fit_transform(X_cluster)

# Apply KMeans with k=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_cluster_scaled)

# Cluster summary table
print("\nCluster Summary:")
cluster_summary = df.groupby('Cluster')[features_cols + [aqi_col]].mean()
cluster_summary['Interpretation'] = ['Low Pollution', 'Medium Pollution', 'High Pollution']
display(cluster_summary)

# Cluster counts
print("\nCluster Counts:")
print(df['Cluster'].value_counts().sort_index())

print("\nDo clusters represent meaningful pollution groups?")
print("Yes! Each cluster shows different average AQI levels representing low, medium, and high pollution.")

# ===== PCA =====
print("\n===== PCA VISUALIZATION =====")

# Apply PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_cluster_scaled)

explained = pca.explained_variance_ratio_
print(f"PC1 Variance Explained: {explained[0]*100:.2f}%")
print(f"PC2 Variance Explained: {explained[1]*100:.2f}%")
print(f"Total Variance Explained: {(explained[0]+explained[1])*100:.2f}%")

# PCA DataFrame
pca_df = pd.DataFrame({
    'PC1': X_pca[:, 0],
    'PC2': X_pca[:, 1],
    'AQI Category': df['AQI Category'],
    'Cluster': df['Cluster'].astype(str)
})

# Plot PCA colored by AQI Category
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

categories = pca_df['AQI Category'].unique()
colors1 = plt.cm.Set1(np.linspace(0, 1, len(categories)))
for cat, color in zip(categories, colors1):
    mask = pca_df['AQI Category'] == cat
    axes[0].scatter(pca_df[mask]['PC1'], pca_df[mask]['PC2'],
                    label=cat, alpha=0.5, s=10, color=color)
axes[0].set_title('PCA colored by AQI Category')
axes[0].set_xlabel(f'PC1 ({explained[0]*100:.1f}%)')
axes[0].set_ylabel(f'PC2 ({explained[1]*100:.1f}%)')
axes[0].legend(fontsize=7, loc='best')

# Plot PCA colored by Cluster
colors2 = ['red', 'blue', 'green']
for cluster, color in zip(['0', '1', '2'], colors2):
    mask = pca_df['Cluster'] == cluster
    axes[1].scatter(pca_df[mask]['PC1'], pca_df[mask]['PC2'],
                    label=f'Cluster {cluster}', alpha=0.5, s=10, color=color)
axes[1].set_title('PCA colored by K-Means Cluster')
axes[1].set_xlabel(f'PC1 ({explained[0]*100:.1f}%)')
axes[1].set_ylabel(f'PC2 ({explained[1]*100:.1f}%)')
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.savefig('pca_visualization.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nPCA helped visualize pollution patterns in 2D!")
print("Points of similar pollution levels are grouped together.")

# Final comparison table
print("\n===== FINAL MODEL COMPARISON =====")
print(f"{'Method':<20} {'Type':<25} {'Result'}")
print("-" * 65)
print(f"{'KNN':<20} {'Supervised':<25} Accuracy = {best_acc:.4f}")
print(f"{'Naive Bayes':<20} {'Supervised':<25} Accuracy = {nb_acc:.4f}")
print(f"{'K-Means':<20} {'Unsupervised':<25} Clusters = 3")
print(f"{'PCA':<20} {'Dimensionality Reduction':<25} Variance = {(explained[0]+explained[1])*100:.2f}%")
```
